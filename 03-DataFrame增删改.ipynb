# DataFrame添加行

```python
# 加载数据集
weather = pd.read_csv('./data/weather.csv')
weather
```

```python
new_series = pd.Series(['高雄市', '南瓜区', '北山', 8.8, 58.0, 0.0], index=['城市','行政区','观测站','气温(度)','相对湿度(%)','累积雨量(mm)'])
weather_df = weather.append(new_series, ignore_index=True)
weather_df
```
